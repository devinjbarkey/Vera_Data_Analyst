# Data Visualization of Capital Punishment Data for 2018

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
file_path = "/Users/devinbarkey/Documents/Projects/cp18st/cp18stat05.csv"
dr_by_race = pd.read_csv(file_path, index_col = 0, skiprows = 10, skipfooter = 3, 
                         thousands = ',', engine = 'python')
```

```python
dr_by_race
```
